```python
df1=df.copy(deep=True)
df1
```

```python
df1.isnull().sum()
```

```python
data=df1.drop(df[['id']],axis=1)
```

```python
data['bmi'].fillna(data['bmi'].median(),inplace=True)
```

```python
data['bmi'].dropna(inplace=True)
```

```python
gender={'Female':0,'Male':1,'Other':2}
data['gender']=data['gender'].map(gender)

smoking_status={'never smoked':0,'Unknown':1,'formerly smoked':2,'smokes':3}
data['smoking_status']=data['smoking_status'].map(smoking_status)

work_type={'Private':0,'Self-employed':1,'children':2,'Govt_job':3,'Never_worked':4}
data['work_type']=data['work_type'].map(work_type)

Residence_type={'Urban':0,'Rural':1}
data['Residence_type']=data['Residence_type'].map(Residence_type)

ever_married={'Yes':0,'No':1}
data['ever_married']=data['ever_married'].map(ever_married)
```

```python
X=data.drop('stroke',axis=1)
y=data['stroke']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=40)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler().fit(X_train)
```

```python
X_train_std=std.transform(X_train)
X_test_std=std.transform(X_test)
```

```python
X_train=X_train_std
X_test=X_test_std
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
predict_train = rfc.predict(X_train)
print('\nTarget on train data',predict_train)
```

```python
predict_test = rfc.predict(X_test)
print('\nTarget on test data',predict_test)
```

```python
accuracy_train = accuracy_score(y_train,predict_train)
print('\naccuracy_score on train dataset : ', accuracy_train)
```

```python
accuracy_test = accuracy_score(y_test,predict_test)
print('\naccuracy_score on test dataset : ', accuracy_test)
```

```python
rfc_cr=(metrics.classification_report(y_test,predict_test))
print(rfc_cr)
```

```python
rcf_cm=confusion_matrix(y_test,predict_test)
rcf_cm
```

```python
n_estimators = [100,140,145,150,160, 170,175,180,200];
cv = StratifiedKFold(n_splits=10,random_state=42,shuffle=True)
```

```python
parameters = {'n_estimators':n_estimators,
              'criterion':['gini','entropy'],
              'max_features':['log2','sqrt'],
              'random_state':[42,12,100,200,35]
              
        }
```

```python
grid = GridSearchCV(estimator=rfc,param_grid=parameters,
                                 cv=cv,
                                 n_jobs = -1)
```

```python
grid.fit(X_train,y_train)
```

```python
print (grid.best_score_)
print (grid.best_params_)
print (grid.best_estimator_)
```
